```python
import matplotlib.pyplot as plt
import torchvision
import torch
import torch.nn as nn
from torchvision import transforms, datasets


trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor()) # datasets.MNIST() 함수는 MNIST 데이터셋을 다운로드해주는 함수, root: 데이터셋이 저장될 경로, train: 학습용 데이터셋 여부, download: 데이터셋 다운로드 여부, transform: 데이터셋에 적용할 변환
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=0)
# batch_size: 배치 크기(64개만큼 샘플링을 해라), shuffle: 매 에폭마다 데이터를 섞을지 여부, num_workers: 데이터를 읽어올 서브 프로세스 수

dataiter = iter(trainloader) # iter() 함수는 반복 가능한 객체를 입력받아 반복자를 반환

try:
    images, labels = next(dataiter) # next() 함수는 반복자를 입력받아 그 다음에 나올 원소를 반환
except StopIteration: # 반복자가 끝났을 때 발생하는 예외
    pass

image_grid = torchvision.utils.make_grid(images) # make_grid() 함수는 입력받은 이미지를 격자 모양으로 만들어주는 함수

plt.imshow(image_grid.numpy().transpose((1,2,0))) # imshow() 함수는 이미지를 출력해주는 함수, transpose() 함수는 행렬의 차원을 바꿔주는 함수, numpy() 함수는 텐서를 넘파이 배열로 변환해주는 함수
plt.axis('off') # axis() 함수는 축의 속성을 설정해주는 함수
plt.show()
```
