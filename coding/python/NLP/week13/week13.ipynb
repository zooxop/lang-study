```python
import matplotlib.pyplot as plt
import torchvision
import torch
import torch.nn as nn
from torchvision import transforms, datasets


trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor()) # datasets.MNIST()  MNIST   , root:   , train:   , download:   , transform:   
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=0)
# batch_size:  (64  ), shuffle:     , num_workers:     

dataiter = iter(trainloader) # iter()       

try:
    images, labels = next(dataiter) # next()        
except StopIteration: #     
    pass

image_grid = torchvision.utils.make_grid(images) # make_grid()       

plt.imshow(image_grid.numpy().transpose((1,2,0))) # imshow()    , transpose()     , numpy()      
plt.axis('off') # axis()     
plt.show()
```
